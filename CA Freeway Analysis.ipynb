```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
loc = 'road_data.csv'
roads = pd.read_csv(loc)
roads = roads.head(30)
norcal = ['Alameda','Santa Clara','San Mateo', 'Contra Costa']
socal = ['Los Angeles','San Diego','Orange']
```

```python
print('Total:')
roads['Vehicle Delay Amount'].sum()
```

```python
print('NorCal Total:')
roads[roads['County'].isin(norcal)].groupby(by='County').sum()['Vehicle Delay Amount'].sum()
```

```python
print('SoCal Total:')
roads[roads['County'].isin(socal)].groupby(by='County').sum()['Vehicle Delay Amount'].sum()
```
